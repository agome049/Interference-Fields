```python
import pygame, sys, time
from pygame.locals import *
import numpy as np
import math

# set up pygame
pygame.init()

# set up the window
WW = 400
WH = 400
windowSurface = pygame.display.set_mode((WW, WH), 0, 32)
pygame.display.set_caption('Animation')

m1,m2,m3,m4 = 1,10,1,1
g= 10
dt=10000

# set up the colors
BLACK = (0, 0, 0)
RED = (255, 0, 0)
GREEN = (0, 255, 0)
BLUE = (0, 0, 255)

#initial positions, velocities, and accelerations of particles
cx1,cy1 =175,175
cx2,cy2 =200,200
cx3,cy3 =175,225
cx4,cy4 =235,200

vx1,vy1 =0,0
vx2,vy2 =0,0
vx3,vy3 =0,0
vx4,vy4 =0,0

ax1,ay1 =0,0
ax2,ay2 =0,0
ax3,ay3 =0,0
ax4,ay4 =0,0

def distance(a,b,c,d):
    return((a-b)**2+(c-d)**2)**.5

def cosine(a,b,c,d):
    return((a-b)/distance(a,b,c,d))

def sine(a,b,c,d):
    return((c-d)/distance(a,b,c,d))

# run the game loop
while True:
    # check for the QUIT event
    for event in pygame.event.get():
        if event.type == QUIT:
            pygame.quit()
            sys.exit()
            
    # draw the black background onto the surface
    windowSurface.fill(BLACK)
    
        # draw the block onto the surface
    for i in range(WW):
        for j in range(WH):
            
            #figure distances from each node
            d1 = ((cx1-i)**2+(cy1-j)**2)**.5
            d2 = ((cx2-i)**2+(cy2-j)**2)**.5
            d3 = ((cx3-i)**2+(cy3-j)**2)**.5
            d4 = ((cx4-i)**2+(cy4-j)**2)**.5
            
            #multliply amplitude to fit in graph
            i1 =255*np.sin(d1/5)
            i2 =255*np.sin(d2/5)
            i3 =255*np.sin(d3/5)
            i4 =255*np.sin(d4/4)
            inter = (i1 + i2 + i3 + i4)/4
            
            #draw interference pattern
            if d1**2<=1 or d2**2<=1 or d3**2<=1 or d4**2<=1:
                windowSurface.set_at([i,j],(0,255,0))
            elif d1!=0 and d2!=0 and d3!=0 and d4!=0:
                if inter >= 0:
                    windowSurface.set_at([i,j],(inter,250/d1**2+250/d2**2+250/d3**2+255/d4**2,0))
                if inter < 0:
                    windowSurface.set_at([i,j],(0,250/d1**2+250/d2**2+250/d3**2+255/d4**2,-inter))
            
            else:
                if inter >= 0:
                    windowSurface.set_at([i,j],(inter,0,0))
                if inter <  0:
                    windowSurface.set_at([i,j],(0,0,-inter))
            
        
    #update position of particles
    cx1 += vx1
    cy1 += vy1
    
    cx2 += vx2
    cy2 += vy2
    
    cx3 += vx3
    cy3 += vy3
    
    cx4 += vx4
    cy4 += vy4
    
    # +accelerations
    if distance(cx1,cx2,cy1,cy2) > 25:
        a1_2=g/distance(cx1,cx2,cy1,cy2)**2
    else:
        a1_2=-g/distance(cx1,cx2,cy1,cy2)**2
    if distance(cx2,cx3,cy2,cy3) > 25:
        a2_3=g/distance(cx2,cx3,cy2,cy3)**2
    else:
        a2_3=-g/distance(cx2,cx3,cy2,cy3)**2
    if distance(cx2,cx4,cy2,cy4) > 25:
        a2_4=g/distance(cx2,cx4,cy2,cy4)**2  
    else:
        a2_4=-g/distance(cx2,cx4,cy2,cy4)**2 
    # -accelerations
    if distance(cx1,cx3,cy1,cy3) > 1:
        a1_3=-10*g/distance(cx1,cx3,cy1,cy3)**2
    
    if distance(cx1,cx4,cy1,cy4) > 1:
        a1_4=-10*g/distance(cx1,cx4,cy1,cy4)**2
    
    if distance(cx3,cx4,cy3,cy4) > 1:
        a3_4=-10*g/distance(cx3,cx4,cy3,cy4)**2
        

    ax1 = m2*a1_2*cosine(cx2,cx1,cy2,cy1)+m3*a1_3*cosine(cx3,cx1,cy3,cy1)+m4*a1_4*cosine(cx4,cx1,cy4,cy1)
    ay1 = m2*a1_2*  sine(cx2,cx1,cy2,cy1)+m3*a1_3*  sine(cx3,cx1,cy3,cy1)+m4*a1_4*  sine(cx4,cx1,cy4,cy1)
        
    ax2 = m1*a1_2*cosine(cx1,cx2,cy1,cy2)+m3*a2_3*cosine(cx3,cx2,cy3,cy2)+m4*a2_4*cosine(cx4,cx2,cy4,cy2)
    ay2 = m1*a1_2*  sine(cx1,cx2,cy1,cy2)+m3*a2_3*  sine(cx3,cx2,cy3,cy2)+m4*a2_4*  sine(cx4,cx2,cy4,cy2)
        
    ax3 = m1*a1_3*cosine(cx1,cx3,cy1,cy3)+m2*a2_3*cosine(cx2,cx3,cy2,cy3)+m4*a3_4*cosine(cx4,cx3,cy4,cy3)
    ay3 = m1*a1_3*  sine(cx1,cx3,cy1,cy3)+m2*a2_3*  sine(cx2,cx3,cy2,cy3)+m4*a3_4*  sine(cx4,cx3,cy4,cy3)
    
    ax4 = m1*a1_4*cosine(cx1,cx4,cy1,cy4)+m2*a2_4*cosine(cx2,cx4,cy2,cy4)+m3*a3_4*cosine(cx3,cx4,cy3,cy4)
    ay4 = m1*a1_4*  sine(cx1,cx4,cy1,cy4)+m2*a2_4*  sine(cx2,cx4,cy2,cy4)+m3*a3_4*  sine(cx3,cx4,cy3,cy4)
    
    vx1 += ax1 - .1*vx1
    vy1 += ay1 - .1*vy1
    
    vx2 += ax2 - .1*vx2
    vy2 += ay2 - .1*vy2
    
    vx3 += ax3 - .1*vx3 
    vy3 += ay3 - .1*vy3
    
    vx4 += ax4 - .1*vx4
    vy4 += ay4 - .1*vy4

    
    # draw the window onto the screen
    pygame.display.update()
    time.sleep(0.02)
```
