```python
import numpy as np
import math

# Constants
M = 100# grid squares per side
r1 =5
r2 =1

# arrays
phi = np.zeros([M+1,M+1],float) # initialize
phi[0,:] = 0

coordx1 = []
coordy1 = []

coordx2 = []
coordy2 = []

for i in range(M+1):
    for j in range(M+1):
        if (i-M/2)**2+(j-M/2)**2 <= r1**2:
            phi[i,j] = 0
            coordx1.append(i)
            coordy1.append(j)
        if (i-M/4)**2+(j-M/4)**2 <= r2**2:
            phi[i,j] = 0
            coordx2.append(i)
            coordy2.append(j)
            
        
xprime = np.array(coordx1)
yprime = np.array(coordy1)

xdime = np.array(coordx2)
ydime = np.array(coordy2)
         
    
    
for i in range(M+1):
    for j in range(M+1):
        
        if (i-M/2)**2+(j-M/2)**2 > r1**2\
        and (i-M/4)**2+(j-M/4)**2> r2**2:
            
            for length in range (xprime.size):
                d1 = ((xprime[length]-i)**2+(yprime[length]-j)**2)**.5
                phi[i,j] = phi[i,j]+np.sin(d1)/d1**2
                
            for length in range (xdime.size):
                d2 = ((xdime[length]-i)**2+(ydime[length]-j)**2)**.5
                phi[i,j] = phi[i,j]+np.sin(d2)/d2**2
```

```python

import matplotlib.pyplot as plt

xx, yy = np.meshgrid(np.arange(M+1),np.arange(M+1))
plt.pcolormesh(xx,yy,phi)
plt.colorbar()
plt.show()
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt


S = 100
m1 = 1
m2 = 1
dt = 1
g = 10
L = 50

vx1,vy1 = (0,0)

vx2,vy2 = (0,0)



def distance(a,b,c,d):
    return((a-b)**2+(c-d)**2)**.5

def cosine(a,b,c,d):
    return((c-d)/distance(a,b,c,d))

def sine(a,b,c,d):
    return((a-b)/distance(a,b,c,d))


psi = np.zeros([S+1,S+1],float)
psi[0,:] = 0

px1,py1 = 50,50
px2,py2 = 20,20

a1 = (g)/distance(px1,px2,py1,py2)
a2 = (g)/distance(px1,px2,py1,py2)

psi[px1,py1] = m1
psi[px2,py2] = m2

vax1,vay1,vax2,vay2 = vx1,vy1,vx2,vy2


for i in range(L):
     
    
    #replaces previous positions of particles
    psi[px1,py1] = 0
    psi[px2,py2] = 0
    
    #updates current positions of particles
    print(vx1,vy1)
    print(vx2,vy2)
    print(vax1,vay1)
    print(vax2,vay2)
    
    px1 += dt*vx1
    py1 += dt*vy1
    px2 += dt*vx2
    py2 += dt*vy2
    
    #updates velocities
    if distance(px1,px2,py1,py2) != 0:
        
        vax1 += -dt*a1*cosine(px1,px2,py1,py2)
        vay1 += -dt*a1*sine(px1,px2,py1,py2)
        vax2 += dt*a2*cosine(px1,px2,py1,py2)
        vay2 += dt*a2*sine(px1,px2,py1,py2)
    
    vx1 =vax1 -vax1%1
    vy1 =vay1 -vay1%1
    vx2 =vax2 -vax2%1
    vy2 =vay2 -vay2%1
    
    vx1=int(vax1)
    vy1=int(vay1)
    vx2=int(vax2)
    vy2=int(vay2)
    
    psi[px1,py1] = m1
    psi[px2,py2] = m2
    
    nn, mm = np.meshgrid(np.arange(S+1),np.arange(S+1))
    plt.pcolormesh(nn,mm,psi)
    plt.colorbar()
    plt.show()
    print(px1,py1)
    print(px2,py2)


```
